# Question No 01

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank_tag=[]
rank_t=driver.find_elements("xpath" , '//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[1]')
for i in rank_t:
    rank=i.text
    rank_tag.append(rank)
rank_tag[:30]
```

```python
name_tag=[]
name_t=driver.find_elements("xpath" , '//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[2]')
for i in name_t:
    name=i.text
    name_tag.append(name)
name_tag[:30]
```

```python
artist_tag=[]
artist_t=driver.find_elements("xpath" , '//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[3]')
for i in artist_t:
    artist=i.text
    artist_tag.append(artist)
artist_tag[:30]
```

```python
upload_date=[]
upload_d=driver.find_elements("xpath" , '//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[5]')
for i in upload_d:
    upload=i.text
    upload_date.append(upload)
upload_date[:30]
```

```python
views_tag=[]
views_t=driver.find_elements("xpath" , '//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[4]')
for i in views_t:
    views=i.text
    views_tag.append(views)
views_tag[:30]
```

```python
print(len(rank_tag),len(name_tag),len(artist_tag),len(upload_date),len(views_tag))
```

```python
you_tube=pd.DataFrame({})
you_tube['Rank']=rank_tag
you_tube['Name']=name_tag
you_tube['Artist']=artist_tag
you_tube['Upload Date']=upload_date
you_tube['Views']=views_tag
```

```python
name_tag=[]
name_t=driver.find_elements("xpath" , "//h3[@class='lister-item-header']//a")
for i in name_t:
    name=i.text
    name_tag.append(name)
name_tag
```

```python
run_time=[]
run_t=driver.find_elements("xpath" , "//span[@class='runtime']")
for i in run_t:
    time=i.text
    run_time.append(time)
run_time
```
